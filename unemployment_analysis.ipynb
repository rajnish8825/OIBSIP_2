```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

```

```python


data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/unemployment.csv")

```

```python


data.head()

```

```python
import pandas as pd
sorted_data = data.sort_values("Date")
region_data = data[data["Region"] == "desired_region"]
average_unemployment = data.groupby("States")["Estimated Unemployment Rate"].mean()

start_date = "2023-01-01"
end_date = "2023-06-30"
date_range_data = data[(data["Date"] >= start_date) & (data["Date"] <= end_date)]
data["Unemployment Rate (%)"] = data["Estimated Unemployment Rate"] * 100
state_totals = data.groupby("States").agg({"Estimated Employed": "sum", "Estimated Unemployment Rate": "sum"})
data = data.drop(["longitude", "latitude"], axis=1)
```

```python


X = data[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate',
       'longitude']]

y = data['Estimated Employed']

```

```python


from sklearn.model_selection import train_test_split

```

```python


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40)

```

```python


X_train

```

```python


from sklearn.linear_model import LinearRegression

```

```python


lm = LinearRegression()

```

```python
lm.fit(X_train, y_train)
```

```python
coeff_data = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
```

```python
coeff_data
```

```python
predictions = lm.predict(X_test)
plt.scatter(y_test, predictions)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.show()
```

```python
import seaborn as sns

sns.histplot((y_test - predictions), bins=50)
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Distribution of Residuals")
plt.show()
```
